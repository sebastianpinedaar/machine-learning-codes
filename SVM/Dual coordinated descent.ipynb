```python
import numpy as np
```

```python
X = np.array([[1.0,0.,2.],
              [1.,2.,0.],
              [1.,2,2.],
              [1.,1.,1.]])

y = np.array([[1],[-1],[1],[-1]])
alpha = [0, 0,0,0]
beta = np.array([0,0,0])
N=X.shape[0]
lamb = 1
ref_clip = 1./(lamb*N)
```

```python
Q= np.multiply(y@y.T, X@X.T)
Q
```

```python
n_iter = 100

for t in range(n_iter):
    print("-----Iteration:", t)
    for n in range(X.shape[0]):
        print("--Sample:", n)
        y_hat = beta.T@X[n,].T
        print("y_hat:", y_hat)
        
        alpha_ = alpha[n] - ((y[n]*y_hat)-1)/Q[n,n]
        print("alpha: ", alpha_)
        
        alpha_ = max(0, min(ref_clip, alpha_))
        print("clipped alpha:", alpha_)
        
        beta = beta + (alpha_-alpha[n])*y[n]*X[n,].T 
        alpha[n] = float(alpha_)
        print("alpha_", alpha_)
        print("beta:", beta)
        print("alpha:", alpha)
        
```
